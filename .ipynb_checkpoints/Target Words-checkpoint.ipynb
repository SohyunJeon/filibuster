```python
import math
from textblob import TextBlob as tb

def tf(word, blob):
    return blob.words.count(word) / len(blob.words)

def n_containing(word, bloblist):
    return sum(1 for blob in bloblist if word in blob)

def idf(word, bloblist):
    return math.log(len(bloblist) / (1 + n_containing(word, bloblist)))

def tfidf(word, blob, bloblist):
    return tf(word, blob) * idf(word, bloblist)
```

```python
bloblist = [tb(regular), tb(regular2), tb(fili)]
for i, blob in enumerate(bloblist):
    print("Top words in document {}".format(i + 1))
    scores = {word: tfidf(word, blob, bloblist) for word in blob.words}
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    for word, score in sorted_words[:3]:
        print("\tWord: {}, TF-IDF: {}".format(word, round(score, 5)))
```

```python
def extract(text):
    result_text = []
    current_name = None
    name_changed = False
    for line in text.split('\n'):
        line = line.strip()
        if not line:
            continue
        if line.startswith('(') and line.endswith(')'):
            continue
        if line.startswith('◯'):
            name_changed = True
#             print(line)
            current_name = re.findall(r'◯(\w+) ', line)[0]
            line = line[line.find('  '):].strip()
        if current_name is None:
            continue
        if name_changed == True:
            result_text = []
            name_changed = False
        result_text.append(line)
    return result_text
```

```python
import math

def tf(word, text):
    return Counter(text)[word] / len(text)

def idf(word, corpus):
    num_containing = sum(1 for text in corpus if word in text)
    return math.log(len(corpus) / (1 + num_containing))

def tfidf(word, text, corpus):
    return tf(word, text) * idf(word, corpus)
```

```python
from collections import defaultdict
```

```python
def count(text):
    text_set = defaultdict(int)
    for word in text.split():
        for target in target_sorted:
            if re.findall(target, word):
                if target in ['국가정보원', '국정원']:
                    target = '국정원'
                if target in ['국가정보원장', '국정원장']:
                    target = '국정원장'
                text_set[target] += 1
#                 print(target, word)
                break
    return text_set
```

```python
import math

def tf(word, counter):
    return counter[word]

def idf(word, corpus):
    num_containing = sum(1 for counter in corpus if counter[word] > 0)
    return math.log(len(corpus) / (1 + num_containing))

def tfidf(word, counter, corpus):
    return tf(word, counter) * idf(word, corpus)
```
